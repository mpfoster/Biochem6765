```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Visualizing the Gibbs-Helmholtz relation
Consider the thermodynamics of binding of M and L to form ML, with a free energy change $\Delta$G, and enthalpy and entropy changes $\Delta$H and $\Delta$S. If the reaction has an associated change in heat capacity, $\Delta$Cp, how do the thermodynamic terms vary with temperature? 

First, let's define some parameters and a temperature range to sample.

```
# visualizing thermodynamic relations
# define temperature points, parameters
t = np.linspace(273,373, num=50)  # temperature range 0-100°C, 50 samples
dCp = -500 # cal/mol/K; assuming temperature independent
T_ref = 298 # K
dh_ref = -5000  # cal/mol; this is dH at T_ref
ds_ref = 50  # cal/mol/K; dS at T_ref
```

## Enthalpy changes linearly with Cp
From the definition Cp = dH/dT, so ΔCp = d$\Delta$H/dT; the integrated form of which is $\Delta H(T) = \Delta H_{ref} + \Delta C_p (T - T_{ref})$

```
# dH(T) = dH_ref + dCp(T-T_ref)
dh = dh_ref + dCp*(t - T_ref)
plt.plot(t,dh)
plt.xlabel('T'); plt.ylabel('Energy')
```

## Entropy changes log-linearly with Cp
From the definition $\Delta C_p = Td \Delta S/dT$, or $d \Delta S = \Delta {C_p dT}/T$; the integrated form of which yields $T\Delta S(T) = T (\Delta S_{ref} + \Delta C_p ln(T/T_{ref}))$

```
# For this purpose, we assume dCp is invariant w. T
# TdS(T) = T(dS_ref + dCp*ln(T/T_ref)
Tds = t*(ds_ref + dCp*np.log(t/T_ref))
plt.plot(t,Tds)
plt.xlabel('T'); plt.ylabel('Energy')
```

## And, $\Delta$G = $\Delta$H - T$\Delta$S:

```
# dG = dH - TdS
dG = dh - Tds
plt.plot(t, dG)
plt.xlabel('T'); plt.ylabel('Energy')
```
