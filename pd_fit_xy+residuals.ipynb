# Generic Python script for plotting and fitting xy data 
Requires pandas, numpy, matplotlib and scipy.optimize

```python
# Generic x/y plotting and fitting in Python; MPF 2022-01
#
# Import commonly used libraries; this should not need edits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

# Read data file using pandas
fname = 'file.csv' # specify here the file name, or URL
fname = 'https://raw.githubusercontent.com/mpfoster/Biochem6765/master/data/binding-data-1.csv'

# this might need modified based on format of input file
df = pd.read_csv(fname, comment = '#', skiprows=None, nrows=None)
df.plot(style='o')
plt.title(fname)
plt.show()
# specify x and y columns:
xcol = 0 # by default, assume first column is the x values
ycol = 2 # columns w/ y values

xval = df.iloc[:,xcol]
yval = df.iloc[:,ycol]

# These commands convert to lists to arrays for doing math on them
xval = np.array(xval); yval = np.array(yval)
# Optionally, specify math on xval, yval; e.g., np.log(xval), 1/yval, 10*xval
x = xval; y = yval 


'''
Now, specify the function to be modeled and fit.
for example, for a straight line for the form y = m*x + b, 
we would need two parameters, m and b, and the independent variable x.
So, the function definition would look like this:
def myfunc(x,m,b):
  return x*m + b

We can use myfunc by giving values of x, m and b; for example:
y = myfunc(1,2,1)
print(y) # would print the result (3, in this case)
(x can be an array of numbers, generating an array of y values)
'''

# function definition
def myfunc(x, Kd):
  return x/(x+Kd) # this is the binding hyperbolic

vars = 1 # initial values of variables for myfunc

def myfunc(x, Kd,n):
  return x**n/(x**n+Kd**n) # this is the Hill eq

vars = 1,1 # initial values of variables for myfunc

# optional customizations:
# see https://matplotlib.org/ for more info
marker = 'o-' # marker '' for default; 'o-' for circle-dash, '--', '..'
xlab = 'x-axis' # change the text in quotes per your data
ylab = 'y-axis' # change the text in quotes per your data
title = 'Title' # change the text in quotes per your data
style = 'ggplot' # specify plot style here; examples at
fontsize = 16 # font size for title and axes
figsize = (6, 6) # width, height
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# if high-resolution plot is desired:
plotfile = '' # options are 'pdf', 'png' or 'svg'; blank just prints to screen

# should not need to edit items below this line
# --------

# for fitting
popt,pcov = opt.curve_fit(myfunc,x,y, vars)
print('----')
print('Best fit of Data fit to function')
print("popt:", popt)
print("pcov:", pcov)
residual = y - myfunc(x,*popt)

xcalc = np.linspace(0.5*min(x), 1.2*max(x))
ycalc = myfunc(xcalc,*popt)

# plots
plt.style.use(style)

# stacked plots with GridSpec
fig = plt.figure(figsize=figsize)

ax1 = plt.subplot2grid((3, 1), (0, 0), rowspan=2)
ax2 = plt.subplot2grid((3, 1), (2, 0), rowspan=1, sharex=ax1)

# plot data
ax1.plot(x,y,marker, label='data')
# plot best-fit line
ax1.plot(xcalc,ycalc, label='popt: ' + str(popt))
# add labels
ax1.set_title('Data fit to Function', fontsize=fontsize)
ax1.set_ylabel(ylab,fontsize=fontsize)
ax1.legend()

# plot residuals
ax2.stem(x,residual, use_line_collection=True)
# add labels
ax2.set_xlabel(xlab,fontsize=fontsize)
ax2.set_ylabel('Residuals',fontsize=fontsize)

fig.tight_layout()
fig.align_ylabels()

# save output?
if plotfile == 'png':
  plt.savefig('plot.png', dpi=300) 
elif plotfile == 'pdf':
  plt.savefig('plot.pdf')
elif plotfile == 'svg':
  plt.savefig('plot.svg')
else:
  plt.show()
```
